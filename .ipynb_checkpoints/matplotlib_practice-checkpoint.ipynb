```python
#import dependencies 
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot- using MATLAB 
plt.plot(x_axis, y_axis)
```

```python
#create the plot- using object oriented interface 
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
#5.1.4 Annotate charts
plt.plot(x_axis,y_axis, marker="*",color="blue",linewidth=2, label='Boston')
#create labels for the x and y axes 
plt.xlabel("Date")
plt.ylabel("Fare($)")
#set the y limit between 0 and 45
plt.ylim(0,45)
#create a title 
plt.title("PyBer Fare by Month")
#add a grid
plt.grid()
#add the legend 
plt.legend()
```

```python
#create a plot with ax.plt()- different option of object oriented interface 
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
ax = plt.axes()
#5.1.4 formatting chart using OOP
ax.plot(x_axis, y_axis, color="green", marker="*", linewidth=2,label='Boston')
#add a legend
ax.legend()
#create a title
ax.set_title("PyBer Fare by Month")
#create axis labels
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
#y axis limit
ax.set_ylim(0,45)
#add gride lines
ax.grid()
```

```python
#5.1.7 Creating Scatter Plots, x=# rides, y= average ride$
x=[50, 42, 39, 34, 32, 48, 22, 17, 8,13]
y=[17, 19, 21, 22, 24, 23, 34, 31, 43, 35]

plt.scatter(x,y, c='green', marker="x",s=x)
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Average Fare vs Total Rides")
plt.show
```

```python
#5.1.8 create pie charts with MATLAB
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow"]
explode=(0,.1,0,0,0,0)
categories=['pants','shirts','outerwear', 'pajamas','underwear','shoes']
revenue=[25,15,5,20,10,25]
plt.subplots(figsize=(8, 8))
plt.pie(revenue,labels=categories, autopct="%1.1f%%",explode=explode,colors=colors)
plt.show
```
